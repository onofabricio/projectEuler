## 10001st prime

By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10 001st prime number?

```python
#contando primos
cont = 0
n = 0
while(cont != 10002):
    flag = False
    if n == 1:
        n += 1
    for i in range(2, n):
        if (n % i) == 0:
            flag = True
            break

    
    if flag:
        pass
    else:
        print(n, "primo ", cont)
        cont+=1
    n += 1
```
