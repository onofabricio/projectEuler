### Number letter counts
#### Problema 17

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
num_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five",
    6: "six",
    7: "seven",
    8: "eight",
    9: "nine",
    10: "ten",
    11: "eleven",
    12: "twelve",
    13: "thirteen",
    14: "fourteen",
    15: "fifteen",
    16: "sixteen",
    17: "seventeen",
    18: "eighteen",
    19: "nineteen",
    20: "twenty",
    30: "thirty",
    40: "forty",
    50: "fifty",
    60: "sixty",
    70: "seventy",
    80: "eighty",
    90: "ninety",
    100: "hundred"
}
```

```python
stringabsurda = ''
for i in range(1, 1001):
    if i in num_dict and i != 100 and i != 1000:
        extenso = num_dict[i]
        stringabsurda += extenso
    elif len(str(i)) == 2:
        extenso = num_dict[int(str(i)[0]+str(0))] + num_dict[int(str(i)[1])]
        stringabsurda += extenso
    elif len(str(i)) == 3:
        extenso = num_dict[int(str(i)[0])]+ 'hundred'
        stringabsurda += extenso
        
        if int(str(i)[-2:]) in num_dict:
            extenso = extenso+ 'and'+num_dict[int(str(i)[-2:])]
            stringabsurda += extenso
        elif str(i)[-2:] == '00':
            pass
        elif str(i)[1] == '0':
            extenso = extenso + 'and' + num_dict[int(str(i)[2])]
            stringabsurda += extenso
        elif str(i)[2] == '0':
            extenso = extenso + 'and' + num_dict[int(str(i)[-2:])]
            stringabsurda += extenso
        
        else:
            extenso = extenso +'and' + num_dict[int(str(i)[1]+str(0))] + num_dict[int(str(i)[2])]
            stringabsurda += extenso
    elif len(str(i)) == 4: 
        extenso = num_dict[int(str(i)[0])]+ 'thousand'
        stringabsurda += extenso
    print(i, extenso, len(extenso))
print(stringabsurda)
len(stringabsurda)
```
