### Digit cancelling fractions
#### Problema 33

The fraction 49/98 is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that 49/98 = 4/8, which is correct, is obtained by cancelling the 9s.

We shall consider fractions like, 30/50 = 3/5, to be trivial examples.

There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.

If the product of these four fractions is given in its lowest common terms, find the value of the denominator.

```python
def cancela(numerador, denominador):
    #print("antes: ", int(numerador),"/", int(denominador))
    for digito in denominador:
        if digito in numerador and digito != 0:
            denominador = denominador.replace(digito, "", 1)
            numerador = numerador.replace(digito, "", 1)
            #print(digito)
            break
    #print("dps: ", int(numerador),"/", int(denominador))
    return int(numerador), int(denominador)
```

```python
for i in range(10,25):
    for j in range(10, i):
        resultado_antes = j/i
        if cancela(str(j), str(i))[1] == 0:
            resultado_dps = resultado_antes + 1
            continue
        resultado_dps = cancela(str(j), str(i))[0] / cancela(str(j), str(i))[1]
        print(j, i, resultado_dps)
        if resultado_antes == resultado_dps:
            print(i, j)
```
